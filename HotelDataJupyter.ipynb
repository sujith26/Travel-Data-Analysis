# Reading TravelRawData.xlsx

```python
df = pd.read_excel(r"C:\Users\hp\Desktop\Data a\TravelRawData.xlsx")
df.head(5)
```

```python
df[['Hotel Name','Hotel location city']][df['Hotel location country']=='US']
```

## Creating two dataframes based on the 'Hotel location country' and 'Customer location country'

```python
df1=df.loc[(df['Hotel location country']=='US') & (df['Customer location - country']!='US')] 
df4=df.loc[(df['Hotel location country']=='US') & (df['Customer location - country']=='US')]
```

## Creating a dataframe with the list of Hotel location cities and the number of bookings done for Hotel Location Cities in the US by customers booking from outside the US. 

```python
df2=df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)
df2
```

## The top 3 Hotel Location Cities in the US for customers booking from outside the US

```python
df3=df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(3)
df3
```

## Finding what share of total bookings to US cities (from customers outside US) do these top 3 cities constitute.

```python
z=(df3['Count'].sum()/df2['Count'].sum())*100
print(z)
print("Top 3 cities constitute "+str("%.2f" % z)+'%'+" of total bookings")
```

## The top 3 Hotel Location Cities in the US for customers booking from inside the US

```python
df5=df4.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(3)
df5
```

## Creating a dataframe with the list of payment methods and the number of payments done through that Payment Type for Hotel Location Cities in the US by customers booking from outside the US.

```python
df6=df1.groupby(['Payment Type'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)
df6
```

```python
df7=df6.head(1)
```

```python
z=(df7['Count'].sum()/df6['Count'].sum())*100
print("The most commonly used Payment Type is "+df6['Payment Type'].iloc[0]+" used by "+str("%.2f" % z)+'%'+" of all users")
```
