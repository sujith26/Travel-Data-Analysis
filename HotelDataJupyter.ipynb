# Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Reading TravelRawData.xlsx

```python
df = pd.read_excel(r"C:\Users\hp\Desktop\Data a\TravelRawData.xlsx")
df.head(5)
```

```python
df[['Hotel Name','Hotel location city']][df['Hotel location country']=='US']
```

## Creating two dataframes based on the 'Hotel location country' and 'Customer location country'

```python
df1=df.loc[(df['Hotel location country']=='US') & (df['Customer location - country']!='US')] 
df2=df.loc[(df['Hotel location country']=='US') & (df['Customer location - country']=='US')]
```

## List of Hotel location cities and the number of bookings done for Hotel Location Cities in the US by customers booking from outside the US. 

```python
df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)
```

## Plot of Top 10 Hotel location cities and the number of bookings done for Hotel Location Cities in the US by customers booking from outside the US.

```python
df3=df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(10)
plt.figure(figsize=(30,10))
plt.barh(df3['Hotel location city'],df3['Count'],color='blue',alpha=0.7,label='Count')
plt.legend(loc='upper right')
plt.xlabel('Number of bookings')
plt.ylabel('Hotel location cities')
plt.rcParams.update({'font.size': 18})
plt.show()
```

## The top 3 Hotel Location Cities in the US for customers booking from outside the US

```python
df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(3)
```

## Finding what share of total bookings to US cities (from customers outside US) do these top 3 cities constitute.

```python
x=df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(3)['Count'].sum()
y=df1.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)['Count'].sum()
z=(x/y)*100
print(z)
print("Top 3 cities constitute "+str("%.2f" % z)+'%'+" of total bookings")
```

## Plot of Top 10 Hotel location cities and the number of bookings done for Hotel Location Cities in the US by customers booking from inside the US.

```python
df4=df2.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(10)
plt.figure(figsize=(30,10))
plt.barh(df4['Hotel location city'],df4['Count'],color='green',alpha=0.7,label='Count')
plt.legend(loc='upper right')
plt.xlabel('Number of bookings')
plt.ylabel('Hotel location cities')
plt.rcParams.update({'font.size': 18})
plt.show()
```

## The top 3 Hotel Location Cities in the US for customers booking from inside the US

```python
df2.groupby(['Hotel location city'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(3)
```

## List of payment methods and the number of payments done through that Payment Type for Hotel Location Cities in the US by customers booking from outside the US.

```python
df1.groupby(['Payment Type'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)
```

```python
x=df1.groupby(['Payment Type'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False).head(1)['Count'].sum()
y=df1.groupby(['Payment Type'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)['Count'].sum()
z=(x/y)*100
print("The most commonly used Payment Type is "+df1.groupby(['Payment Type'])['Cookie ID'].count().reset_index(name="Count").sort_values('Count',ascending=False)['Payment Type'].iloc[0]+" used by "+str("%.2f" % z)+'%'+" of all users")
```
